```python
import os
import cv2

# Load the class names from the classes.txt file
with open("1/classes.txt", "r") as classes_file:
    class_names = classes_file.read().strip().split("\n")

# Function to read YOLO format annotation file and display annotations on the image
def show_annotations(image_path, annotation_path):
    try:
        image = cv2.imread(image_path)
        if image is None:
            print(f"Corrupted image: {image_path}")
            return

        with open(annotation_path, "r") as annotation_file:
            annotations = annotation_file.readlines()

        for annotation in annotations:
            annotation = annotation.strip().split()
            class_id = int(annotation[0])
            x, y, w, h = map(float, annotation[1:])
            class_name = class_names[class_id]
            x1 = int((x - w / 2) * image.shape[1])
            y1 = int((y - h / 2) * image.shape[0])
            x2 = int((x + w / 2) * image.shape[1])
            y2 = int((y + h / 2) * image.shape[0])
            
            # Draw a bounding box and label on the image
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, class_name, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Display the image with annotations
        cv2.imshow("Image with Annotations", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    except Exception as e:
        print(f"Error processing {image_path}: {str(e)}")

# Directory containing images and annotations
data_dir = "1"

# Iterate through the files in the directory
for filename in os.listdir(data_dir):
    if filename.endswith(".jpg"):
        image_path = os.path.join(data_dir, filename)
        annotation_path = os.path.join(data_dir, filename.replace(".jpg", ".txt"))
        
        if not os.path.exists(annotation_path):
            print(f"Annotation file not found for {filename}")
            continue

        show_annotations(image_path, annotation_path)

cv2.destroyAllWindows()  # Close any remaining open windows
```
